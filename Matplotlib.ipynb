__BAR GRAPH__

```python
import matplotlib.pyplot as plt
x = ["part1", "part2", "part3", "part4", "part5"]
y = [96,84,57,62,89]
color = ["red","green", "orange", "yellow", "pink"]
plt.bar(x,y, color = color, edgecolor = "black")
plt.xlabel("Parts of HP" , fontsize = 15)
plt.ylabel("Ratings", fontsize = 15)
plt.title("BAR GRAPH", fontsize = 18)
plt.show()
```

__IMPORT FROM EXCEL FILE BAR GRAPH__

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_excel("payments.xlsx")
df = pd.DataFrame(data)
#print(df)
group = df.groupby("Payment_Method")["Amount"].sum()
print(group)
colors = ("green", "red", "orange")
plt.bar(group.index,group.values, color = colors, edgecolor = "black")
plt.xlabel("Payment_Method", fontsize = 15)
plt.ylabel("Amount", fontsize = 15)
plt.title("BAR Graph", fontsize = 18)
plt.show()
```

__LINE GRAPH__

```python
import matplotlib.pyplot as plt

x = ["Day1", "Day2", "Day3", "Day4", "Day5"]
y = [450,300,250,400,200]
y1 = [200,365,165,456,652]
plt.plot(x,y, marker = "o", label = "week1", color = "green", ls="--", alpha = 0.5)
plt.plot(x,y1, marker = "o", label = "week2", color = "red")
plt.legend()
plt.xlabel("Days", fontsize = 15)
plt.ylabel("Data in y & y1", fontsize = 15)
plt.title("TOTAL EXPENDITURE", fontsize = 18)
plt.show()
```

__SCATTER PLOT__

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(1,10,50)
y = np.random.randint(10,100,50)
color = np.random.randint(10,100,50)
plt.scatter(x,y, cmap = "hot", c= color)
plt.colorbar()
plt.title("SCATTER PLOT")
plt.ylabel("Range")
plt.xlabel("Random value")
plt.show()
```

__STACK PLOT__

```python
import matplotlib.pyplot as plt

Days = [1,2,3,4,5,6,7]

NOP1 = [5,10,30,20,35,60,80]
NOP2 = [10,23,29,36,45,25,64]
NOP3 = [8,30,50,60,70,90,110]
c = ["red", "orange", "green"]
plt.stackplot(Days, NOP1, NOP2, NOP3, colors = c, labels = [ "Week1", "Week2", "Week3"])
plt.legend()
plt.title("STACK PLOT")
plt.xlabel("Days")
plt.ylabel("No. of Persons")
plt.show()
```

__STEP PLOT__

```python
import matplotlib.pyplot as plt

x = ["Day1", "Day2", "Day3", "Day4", "Day5"]
y = [20,35,45,26,29]
plt.step(x,y, marker = "D", where = "post")
plt.title("STEP PLOT")
plt.xlabel("Days")
plt.ylabel("Customers")
plt.show()
```

__Now we done the STEM plotting on the dataset__

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_excel("payments.xlsx")
df = pd.DataFrame(data)
#print(df)
group = df.groupby("Category")["Amount"].sum()
print(group)
plt.step(group.index, group.values, marker = "o", where = "post")
plt.title("Expenses")
plt.xlabel("Category")
plt.ylabel("Amount")
plt.show()
```
